# A Dataframe with 32 observations on 11 (numeric) variables


mpg: Miles/(US) gallon
cyl: Number of cylinders
disp: Displacement (cu.in.)
hp: Gross horsepower
drat: Rear axle ratio
wt: Weight (1000 lbs)
qsec: 1/4 mile time (seconds)
vs: Engine configuration (0 = V-shaped, 1 = straight)
am: Transmission (0 = automatic, 1 = manual)
gear: Number of forward gears
carb: Number of carburetors 

 

```python
mtcars.dtypes
```

```python
#change data types
mtcars.head(3)
```

```python
# Covert selected columns to category
mtcars = data('mtcars')
mtcars = mtcars.astype({'cyl':'category','vs':'category','carb':'category','am':'category','gear':'category'})
```

```python
mtcars.dtypes
```

```python
mtcars.head()
```

```python
mtcars['am'] =mtcars['am'].map({0 :'auto', 1:'manual'})
mtcars.head() #Sorting by manual to automatic engine
```

```python
mtcars['gear'] =mtcars['gear'].map({4 :'munira', 3:'shumi'})
mtcars.head()
```

```python
mtcars['am'].value_counts()
```

```python
mtcars['am'] =mtcars['am'].map({0:'shumi', 1:'munira'})
mtcars.head()
```

```python
mtcars['mpg'].max()
```

```python
mtcars[mtcars['mpg'] == 33.9]
```

```python
mtcars[mtcars['mpg'] > 30]
```

```python
mtcars[mtcars['mpg']>30].sort_values('mpg')
```

```python
mtcars[mtcars['mpg']>30].sort_values('mpg',ascending=False)
```

```python
mtcars[mtcars['mpg'] == mtcars['mpg'].max()]
```

```python
mtcars[mtcars['mpg'] >= mtcars['mpg'].mean()]
```

```python
bins = [0,17,29, mtcars['mpg'].max()]
labels =['low','medium','high']
```

```python
mtcars['mpg_level'] = pd.cut(mtcars['mpg'], bins=bins,labels=labels,include_lowest=True)
```

```python
mtcars.head()
```

```python
mtcars.tail()
```

```python
bins = [0,70,150,250, mtcars['hp'].max()]
labels =['low','medium','high','very high']
mtcars['hp_level'] = pd.cut(mtcars['hp'], bins=bins,labels=labels,include_lowest=True)
```

```python
mtcars['hp'].value_counts()
```

```python
mtcars.head()
```

```python
mtcars.tail()
```

```python
mtcars['gear'].value_counts()
```

```python
mtcars['gear'] =mtcars['gear'].map({3:'shumi', 4:'munira'})
mtcars.head()
```

```python
mtcars.groupby('mpg_level')['mpg'].count()
```
