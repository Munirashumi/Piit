# Graphs With Seaborn Library

```python
import pandas as pd
import numpy as np
from pydataset import data
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = data('mtcars')
df.head()
```

```python
sns.scatterplot(data=df,x='wt',y='mpg',hue='cyl')
plt.show()
```

```python
sns.regplot(data=df,x='hp',y='mpg')
plt.show()
```

```python
sns.pairplot(data=df[['mpg','hp','wt','cyl']])
plt.show();
```
