# Missing Values

```python
import pandas as pd
import numpy as np
from pydataset import data
```

```python
# Create Data
# 4 types of missing values-None,np.nan,pd.NA,NaT
import numpy as np
data = [
    [1,'Taheera', None ,86], 
    [2,'Aelita',56,90],
    [3,'Kulan',np.nan, None ],
    [4,'Batgerel',71,79],
    [5,'Manish', np.nan,78]
]
df = pd.DataFrame(data, columns=['Rollno','Name','Maths','English'])
df
```

```python
# How to find missing values from thousands of data
df.isna().sum()
```

```python
df.isna().sum(axis=1)
```

```python
# How to fill missing values
df.fillna(0)
```

```python
df.fillna(40)
```

```python
df[df['Rollno']==1]
```

```python
df.loc[df['Rollno']==1,['Maths']]
```

```python
df.loc[df['Rollno']==1,['Maths','English']]
```

```python
df.loc[df['Rollno']==1,['Maths']]=88
df
```

```python
df.loc[df['Rollno']==3,['Maths','English']]=[100,100]
df
```

```python
df['Maths'].mean()
```

```python
df['Maths']=df['Maths'].fillna(df['Maths'].mean())
df
```

```python
df['English']=df['English'].fillna(df['English'].min())
df
```

```python
data2 = [
    [1,'Taheera', 77], 
    [2,'Aelita',77],
    [3,'Kulan',77],
    [4,'Batgerel',77],
    [5,'Manish',77]
]
df2 = pd.DataFrame(data2, columns=['Rollno','Name','French'])
df2
```

```python
df3 = pd.merge(df,df2, on=['Rollno','Name'])
df3
```
