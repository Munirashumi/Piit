```python
import matplotlib.pyplot as plt
```

```python
# 1. Histogram
plt.figure(figsize=(15,5))
#plt.hist(df['mpg'], bins=5, edgecolor='black')
plt.hist(df['mpg'],bins=[10,15,20,25,35],edgecolor='black',color='g',orientation='vertical')
plt.title("Distribution of MPG")
plt.title("Distribution of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
# MPG distribution: Mostly clustered between ~18–24 → typical mid-efficiency cars.
```

```python
# 2. Scatter
plt.figure()
plt.scatter(x=df['hp'], y=df['mpg'],c='g',s=20)
plt.title("HP vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.show()
```

```python
# 3. Boxplot
plt.figure()
df.boxplot(column='mpg', by='cyl', patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'))
plt.title("MPG by Cylinder")
plt.xlabel("Mileage")
plt.ylabel("Cylinder")
plt.suptitle("")
plt.show()
```

```python
# 4. Line plot
df_sorted = df.sort_values(by='wt')
plt.figure()
plt.plot(df_sorted['wt'], df_sorted['mpg'])
plt.title("Weight vs MPG")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```
