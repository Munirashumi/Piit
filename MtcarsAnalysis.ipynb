# Part-4 Answer: Data Visualization


```python
import matplotlib.pyplot as plt
```

```python
# Q.A:(12):
df.groupby('cyl')['mpg'].mean().plot(kind='bar')
plt.title("Average MPG by Cylinders")
plt.show()
```

```python
df.plot(kind='scatter', x='wt', y='mpg', title='wt vs MPG')
plt.show()
```

```python
# Scatter
plt.figure()
plt.scatter(df['wt'],df['mpg'])
plt.title("wt vs MPG")
plt.xlabel("wt")
plt.ylabel("MPG")
plt.show()
```

```python
# Histogram
df['hp'].plot(kind='hist',title='Distribution of HP')
plt.show()
```

```python
# Histogram
plt.figure()
plt.hist(df['mpg'])
plt.title("Distribution of HP")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot
df.boxplot(column='mpg', by='am')
plt.title("Mpg by Transmission")
plt.show()
```

```python
# Boxplot
plt.figure()
df.boxplot(column='mpg', by='cyl')
plt.title("MPG by Cylinders")
plt.suptitle(" ")
plt.show()
```
