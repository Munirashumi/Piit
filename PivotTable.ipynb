# Pivot Table

- Index= what goes down the rows
- columns= → What goes across the top
- values= → What fills the table
- aggfunc= → How to summarize if there are duplicates

```python
mtcars.head()
```

```python
#how many cars have gear=5, am=0
```

```python
mtcars.pivot_table(values ='mpg', index='gear', observed=False)
#no aggregation or defalt aggregation is mean
#mean mpg of gear3 car is 16
```

```python
#?mtcars.pivot_table
```

```python
mtcars.pivot_table(values='hp',index='cyl',observed=False)
#mean hp of cyl 4 car is 82
```

```python
mtcars.pivot_table(values='mpg',index=['gear','am'],observed=False,aggfunc=['sum','mean'])
```

```python
mtcars.pivot_table(values=['mpg','hp'],index=['gear','am'],observed=False,aggfunc=['sum','mean','count']).reset_index()
#sum of mpg for each gear type
```

```python
mtcars.pivot_table(values=['mpg','hp'],index=['gear','am'],observed=True,aggfunc=['sum','mean','count']).reset_index()
```

```python
pd.pivot_table( data=mtcars,observed=False,
    index='gear',       #Rows
    values='mpg',        #Values to summerise
    aggfunc='median',    #Aggregation function
)
#median of gear 3 car is 15
```

```python
pivot1 = pd.pivot_table(
    data=mtcars,observed=False,
    index='gear',    #Rows
    columns='am',   #Coloumns
    values='mpg',    #Values to summerise
    aggfunc='mean',  #Aggregation function
)
```

```python
pivot1
# Mean of gear-3 & auto TX -16 while for gear-3 & manual TX- no cars
```

```python
print(mtcars.am.value_counts(),mtcars.cyl.value_counts())
```

```python
pd.pivot_table(observed=False, data=mtcars, index='gear', columns='am', values='mpg', aggfunc=['mean','max','min','count','sum']).reset_index()
# 32Cars
```

```python
pd.pivot_table(data=mtcars, index=['gear','am'], columns='cyl', values='mpg', aggfunc=['mean','max','min','count','sum','std']).reset_index()
#19.75 mean mpg gear-3,am0,cyl6
```

```python
### wide to long
# roll,name,subject,marks
```

```python
df =  pd.DataFrame(data)
```

```python
df_long = df.melt(id_vars=['Rollno','Name'],
                  value_vars=['Maths','English'],
                  var_name='Subject',
                  value_name='Marks')
df_long
```

```python
long_df = df.melt(id_vars=['Rollno','Name'],
                  value_vars=['Maths','English'],
                  var_name='Subject',
                  value_name='Marks')
long_df
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'],columns='Subject',values='Marks').reset_index()
wide_df
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'],columns='Subject',values='Marks')
wide_df
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'],columns='Subject',values='Marks').reset_index(drop=False).rename_axis(None,axis=1)
wide_df
```

```python
wide_df2 = df_long.pivot_table(index=['Rollno','Name'],columns='Subject',values='Marks',aggfunc='sum').reset_index(drop=False).rename_axis(None,axis=1)
wide_df2
```
