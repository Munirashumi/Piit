```python
# 8. Splitting and Joining
print("\n8. Split & Join")
sentence = "Python is easy to learn"
print(' the Sentence is ', sentence)
word = sentence.split(" ")
print("Split:", word)
joined = "-".join(word)
print("joined:",joined)
```

```python
# 9. String Formatting
# -------------------------------
print("\n9. Formatting")
name = "Taheera"
age = 25
gender ='F'
city ='New York'
# Old style
print("My name is %s and I am %d years old, and my gender is %s" % (name, age,gender))
# format()
print("My name is {} and I am {} years old, with gender is {} and city in {}".format(name, age, gender, city))
# f-string (recommended)
print(f"My gender is {gender}, name is {name} and I am {age} years old")
```

```python
# 9. String Formatting
print("\n9. Formating")
name = "Munira"
age = 25
gender = 'F'
city = 'Virginia'
# Old style
print("My name is %s and I am %d years old, and my gender is %s" % (name,age,gender))
# Format()
print("My name is {} and I am {} years old, and my gender is {} and city in {}".format(name,age,gender,city))
# F-string(recommended)
print(f"My gender is{gender},name is {name} and I am {age} years old")
```

```python
import pandas as pd
from pydataset import data
```

```python
# Create mtcars dataset with car names
data = {
    "car_name": [ "Mazda RX4", "Mazda RX4    Wag", "Datsun 710","Hornet 4 Drive", "Hornet   Sportabout  ", "Valiant", "Duster 360", "merc 240D", "Merc 230", "Merc 280" ],
    "mpg": [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8, 19.2]}
df = pd.DataFrame(data)
print("Original Data:\n", df)
```

```python
# Search
mazda_cars = df[df['car_name'].str.contains("Mazda")]
print("n\Cars with Mazda':\n",mazda_cars)
# .str.contains() → pattern search
# Case sensitive by default
```

```python
df[df['car_name'].str.contains("Merc", case=True)]
```

```python
df[df['car_name'].str.contains("merc", case=False)]
# case insenstive
```

```python
# FILTER (Start / End patterns
# Cars starting with “Merc”
merc_cars = df[df['car_name'].str.startswith("Merc")]
print("\nCars starting with 'Merc':\n",merc_cars)
```

```python
df[df['car_name'].str.startswith(("Merc",'merc'))]
```

```python
df[df['car_name'].str.endswith("Drive")]
```

```python
# Replace Strings
df['car_name_updated'] = df['car_name'].str.replace("Merc", "Mercedes", regex=False)
print("\nUpdated_name':\n",df[['car_name','car_name_updated']])
```

```python
df['car_name'].str.replace("Merc", "Mercedes", case=False, regex=False)
```

```python
df['car_name'].str.replace("Merc", "Mercedes", case=True, regex=True)
```

```python
# Split /Breaks
df['brand']= df['car_name'].str.split().str[0]
print("\nExtracted Brand':\n",df[['car_name','brand']])
#👉 Very useful for: * extracting categories, * feature engineering
```

```python
df['upper_name'] = df['car_name'].str.upper()
df['lower_name'] = df['car_name'].str.lower()
df
```

```python
# Cleaning (trim spaces)
df['clean_name'] = df['car_name'].str.strip()
df
```
